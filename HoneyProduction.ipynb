```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv('C:/Users/Mathesh/OneDrive/Desktop/ML/honeyproduction.csv')
df.head()
```

```python
prod_per_year=df.groupby('year').totalprod.mean().reset_index()
print(prod_per_year)
```

```python
X=prod_per_year["year"]
X = X.values.reshape(-1, 1)
print(X)
```

```python
y=prod_per_year["totalprod"]
print(y)
```

```python
regr = linear_model.LinearRegression()
regr.fit(X,y)
```

```python
y_predict=regr.predict(X)
```

```python
X_future = np.array(range(2013, 2051))
X_future = X_future.reshape(-1, 1)
future_predict=regr.predict(X_future)
plt.scatter(X,y)
plt.title('Honey Production')
plt.plot(X,y_predict)
plt.plot(X_future,future_predict)
plt.show()
index_2050 = 2050 - 2013
production_2050 = future_predict[index_2050]
print(f"Predicted honey production in 2050: {production_2050}")
```
